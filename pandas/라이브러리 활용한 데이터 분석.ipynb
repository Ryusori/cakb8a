#### 특잇값을 찾고 제외하기 (p288)

- 배열 연산을 수행할 때는 특잇값을 제외하거나 적당한 값으로 대체하는 것이 중요하다. 

```python
data = pd.DataFrame(np.random.randn(1000, 4))
data.describe()
```

```python
# 절댓값이 3을 초과하는 값을 찾아내자.

col = data[2]
col[np.abs(col) > 3]
```

```python
def place_find(s) :
    # 접속 url
    url = 'http://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(s)
    
    # headers 입력
    headers = {
        # 카카오지도 접속
        "Authorization" : "KakaoAK 4066642da7df95e0189bf6934298ba41"
    }
    # API 요청
    palces = requests.get(url, headers=headers).json()['documents']
    try :
        palces = palces[0]
        jejuGwanGwang1_name  = palces['place_name']
        위도 = palces['y']
        경도 = palces['x']
        
        result = [jejuGwanGwang1_name, 위도, 경도]
        
        return result
    except:
        pass      
```
