### ☆검색어 함수

```python
# 검색어를 넣으면 네이버 뉴스 출력하는 함수 만들기
def search_naver(keyword):
    url1 = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query="
    url2 = keyword
    url3 = "&start="

    for i in range(1, 3):
        url = url1 + url2 + url3 + str(i)
        html = requests.get(url)
        soup = BeautifulSoup(html.text, 'html.parser')
        title = soup.find_all(class_ = "news_tit")
        for j in title:
            print(j.text, ":", j.attrs['href'])
        print(f"{i}페이지 입니다.")
        print("-"*50)
        print() 
```

### ★선생님이 만드신 데이터 프레임 함수

```python
# 네이버 데이터 프레임 함수
import pandas as pd
def search_naver(keyword):
    li1=[]
    li2=[]
    url1 = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query="
    url2 = keyword
    url3 = "&start="

    for i in range(1,3):
        url = url1 + keyword + url3 + str(i)
        html = requests.get(url)
        soup = BeautifulSoup(html.text, 'html.parser')
        title = soup.find_all(class_ = "news_tit")
        for j in title:
            li1.append(j.text)
            li2.append(j.attrs["href"])
        # print(li1)
        # print(li2)
        data = pd.DataFrame({"제목":li1,"링크":li2})
        data.to_csv(".csv")
        return data

        # print(f"{i}페이지 입니다.")
        # print("-"*50)
        # print()    
```

### ☆날씨정보-> json형식
